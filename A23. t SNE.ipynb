# t-SNE
### (Dimensional Reduction Approach) It only used for visualization purpose

- We get more data points in tail region bcz of that we use t distribution

- It calculates the probability between datapoint in High dimensional subspace and Low dimnesional subspace and Aim is to match these probabilities.

- P(j!i) Probability of i that i record will pick the j record as its neighbor 

- It should match Low dimensional data to Hogh dimensional data. 
- If it is matching then move forward with the data visaulizaton.

- KL Divergence = measure of how different one probability distribution from a second

- Gradient Descent = Change the data point  in Lower dimensional subspace, unless and untill KL values is minimum 

- Gradient Descent = It is keep on changing the values unless and until it will reaches the minium KL values 

- tsne is unsupervised approach
- For PSA and t-SNE is DR approach no need of target column 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\sunny\Desktop\DS Material\Data Set\Wine.csv")
df.head()
```

```python
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
x = df.iloc[:,:-1]
x
```

```python
tsne = TSNE(n_components=2) 
# n_components is 2 = 2 columns
score = tsne.fit_transform(x)
# fit= will calu=culate the probability, distance 
# transform = will replace the values
score
```

```python
cluster.tsneplot(score, colorlist= df['Customer_Segment'])
# It is saved in current working directory
```
