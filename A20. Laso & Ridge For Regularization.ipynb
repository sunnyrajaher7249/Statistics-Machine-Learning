```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size= 0.3 , random_state=1)
```

#  Model Building Linear regression


```python
lr = LinearRegression()

lr.fit(xtrain,ytrain)

ypred = lr.predict(xtest)

print('Training Accuracy', lr.score(xtrain,ytrain))
print('Testing Accuracy', lr.score(xtest,ytest))
```

- Training Acc is High = Low Bias
- Testing Acc is low = High variance

## overfitting

###  Model Building Ridge

- Ridge is not going to use for feature selection

```python
rg = Ridge(alpha= 100) # 100 is random value

rg.fit(xtrain,ytrain)

rgpred = rg.predict(xtest)

print('Training Accuracy', rg.score(xtrain,ytrain))
print('Testing Accuracy', rg.score(xtest,ytest))
```

# Lasso
