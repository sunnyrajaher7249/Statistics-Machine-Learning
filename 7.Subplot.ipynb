```python
 x = [1,2,3,4,5]
 y =[5,6,9,8,12,]  

fig,ax = plt.subplots(2,2, figsize =(10,7))
ax[0,0].plot(x,y, color = 'red')
ax[0,1].scatter(x,y)
ax[1,0].boxplot(x,y)
ax[1,1].scatter(x,y, color = 'green')
ax[0,0].set_title('1st')
ax[0,1].set_title('2nd')
ax[1,0].set_title('3rd')
ax[1,1].set_title('4th')
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head(2)
```

```python
pd.set_option('display.max_columns', None)
df.head(2)
```

```python
fig,ax = plt.subplots(2,2, figsize =(15,12))
sns.countplot(df['fuel-type'],ax= ax[0,0])
sns.scatterplot(data= df, x= 'city-mpg', y = 'highway-mpg',ax= ax[0,1])
sns.distplot(df['compression-ratio'],ax = ax[1,0])
sns.boxenplot(df['width'],ax = ax[1,1])

ax[0,0].set_title('1st', color = 'red' , fontsize = 18 , fontweight = 'bold')
ax[0,1].set_title('2st', color = 'red' , fontsize = 18 , fontweight = 'bold')
ax[1,0].set_title('3st', color = 'red' , fontsize = 18 , fontweight = 'bold')
ax[1,1].set_title('4st', color = 'red' , fontsize = 18 , fontweight = 'bold')
```
