# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x,method='single')
dendrogram = hierarchy.dendrogram(lk)

# We use single linkage bcz of that the it is not clearly intrepretable
# Single linkage linkage use minmum distance between datapoint and cluster
```

```python
lk = hierarchy.linkage(x,method='complete')
dendrogram = hierarchy.dendrogram(lk)
plt.axhline(y = 7 , c = 'red', linestyle = '--')
```

# Model Building

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=2)

y = hc.fit_predict(x)
```

```python
y
```

```python
df['Cluster'] = y
df
```

# In-class activity
- Create Logistic regression model.
- Steps:
    1. Define x and y
    2. Split the data into training and testing
    3. Build model
    4. Get prediction
    5. Evaluate model


```python
df.drop(columns=['y'], inplace = True)
```

```python
df
```

```python
x
```

```python
y = df['Cluster']
y
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()

lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```
