# Why to visualize data?
 + Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from. The main goal of data visualization is to make it easier to identify patterns, trends and outliers in large data sets.
   + Better analysis
   + Quick action
   + Identifying patterns
   + Finding errors
   + Understanding the story
   + Exploring business insights
   + Grasping the Latest Trends

```python
x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y, 'r--')
x1 = [2,4,6,8]
y1 = [3,5,7,9]

plt.plot(x1,y1)

plt.xlabel( 'X Axis')
plt.ylabel( 'Y Axis')
plt.title( 'X vs Y Line Graph')

plt.fill_between(x, y, y1, color = 'red', alpha = 1)
```

+ Xlim: used to define limit for x variable

+ Ylim:used to define limit for y variable

```python
Area = [1000,1246,2500,2900,4500,5800]
Price =[10,35,45,65,18,34]
plt.plot(Area, Price)
plt.grid()
plt.grid( color ='red' )

plt.title('House Prising')
plt.xlabel('Area of the house')
plt.ylabel('price of the House')

plt.xlim(500,6000)
```

+ xticks: rotate label
+ yticks: rotate label

```python
Area = [1000,1246,2500,2900,4500,5800]
Price =[10,35,45,65,18,34]
plt.plot(Area, Price)
plt.grid()
plt.grid( color ='red' )

plt.title('House Prising' , fontsize = 16 , fontweight = 'bold', color = 'green')
plt.xlabel('Area of the house')
plt.ylabel('price of the House')

plt.xlim(500,6000)
plt.xticks(rotation =90 ,fontsize = 16 , fontweight = 'bold')
plt.yticks(rotation =45 , fontsize = 16, fontweight = 'bold')
```

+ Multiple Plots : multiple line in one chart called multiple plot

# Matplotlib Histogram
 + A histogram is a graph showing frequency distribution
 
 
 + It is a graph showing the number of observations within each given interval

```python
import pandas as pd
```

```python
df = pd.read_csv('Automobile_data.csv')
```

```python
pd.set_option('display.max_columns' , None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe().T
```

```python
plt.hist(df['width'] , bins= 10)
plt.title('Width frequency Table')
plt.xlabel('Width')
plt.ylabel('Frequency')
```
